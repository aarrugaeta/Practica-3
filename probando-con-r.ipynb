### Crear datos
A través del código de un gráfico proporcionado por el profesor vamos a cambiar los diferentes campos de variables y valores con el objetivo de crear un nuevo gráfico:

```r
# Create a dataset
bajas <- c(rep("Tanques" , 2) , rep("Artillería" , 2) , rep("Aeronaves" , 2) )
paises <- rep(c("Rusia" , "Ucrania") , 3)
value <- abs(rnorm(6 , 0 , 15))
data <- data.frame(bajas,paises,value)

# Grouped
ggplot(data, aes(fill=paises, y=value, x=bajas)) + 
    geom_bar(position="dodge", stat="identity")
```

Con "create a dataset" lo que hacemos es crear un fichero de datos que recoga las variables que a nosotros nos interesan. Con ggplot lo que conseguimos es agrupar estos datos en un gráfico y visualizarlo.

En clase vimos que a cada una de nosotras nos salía un gráfico distinto a pesar de utilizar lo que aparentemente sería el mismo dataset. Esto se debe a que para generar los valores de nuestras variables estamos utilizando la función "rnorm" que lo que hace es generar n números aleatorios en función de una distribución normal. Para que todas pudieramos tener la misma gráfica utilizaríamos la función "set.seed(n)" que fija una semilla a partir de la cual se generan los números aleatorios (todas tendríamos que meter el mismo número).

```r
# Create a dataset
bajas <- c(rep("Tanques" , 2) , rep("Artillería" , 2) , rep("Aeronaves" , 2) )
paises <- rep(c("Rusia" , "Ucrania") , 3)
set.seed(10)
value <- abs(rnorm(6 , 0 , 15))
data2 <- data.frame(bajas,paises,value)

# Grouped
ggplot(data2, aes(fill=paises, y=value, x=bajas)) + 
    geom_bar(position="dodge", stat="identity")
```
